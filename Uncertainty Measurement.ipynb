```python
# Uniform Distrubution

import numpy as np
import random
import matplotlib.pyplot as plt
import statistics
import time
import math
from scipy.special import gamma
plt.style.use('grayscale')
def uniform(a,b,vals):
    l=[]
    for i in range(vals):
        l.append(a+(b-a)*random.random())
    return l

alphas=uniform(9.5*10**-6,13.5*10**-6,100000)

plt.hist(alphas)
plt.show()
```

```python
# Student’s t distribution
# t distribution for input Ls

import numpy as np
import random
import matplotlib.pyplot as plt
import statistics
import time
import math
from scipy.special import gamma
plt.style.use('grayscale')

def t(u,sigma,dof,vals):
    t_dist=[]
    i=0
    r=uniform(0,1,vals)
    
    while i<vals:
        r1=random.choice(r)
        r2=random.choice(r)
        if r1<0.5:
            t=1/(4*r1-1)
            v=r2/(t*t)
        else:
            t=4*r1-3
            v=r2
            
        if v<(1-abs(t)/2) or v< math.pow((1+(t*t)/dof), -(dof+1)/2):
            t_dist.append(t)
            i+=1
        else:
            continue
    temp2=[]
    for i in range(len(t_dist)):
        temp=u+sigma*t_dist[i]
        temp2.append(temp)
    return temp2

Ls=t(50000623*10**(-9),25*10**(-9),18,1000000)
                
plt.hist(Ls,bins=100)
plt.show()              
```

```python
# Arcsine Distribution

import numpy as np
import random
import matplotlib.pyplot as plt
import statistics
import time
import math
from scipy.special import gamma
plt.style.use('grayscale')

def arcsine(a,b,vals):
    r=uniform(0,2*np.pi,vals)
    r=(a+b)/2+((b-a)/2)*np.sin(r)
    return(r)

delta=arcsine(-0.5,0.5,1000000)
plt.hist(delta,bins=100)
plt.show()
```

```python
# CTrap Distribution
# CTrap distribution for input d_theta

import numpy as np
import random
import matplotlib.pyplot as plt
import statistics
import time
import math
from scipy.special import gamma
plt.style.use('grayscale')

def ctrap(a,b,d,vals):
    ctrap=[]
    for i in range(vals):
        aS=(a-d)+2*d*random.random()
        bS=(a+b)-aS
        r=aS+(bS-aS)*random.random()
        ctrap.append(r)
    return ctrap

dtheta=ctrap(-0.05,0.05,0.025,1000000)

plt.hist(dtheta,bins=100)
plt.show()
```
