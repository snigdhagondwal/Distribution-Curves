# Uniform Distribution
Since, area under the curve = 1
In, uniform distribution height is given by the interval of the curve, as follows:

l * b = area
(b-a) * h = 1
        h = 1/ (b-a)

```python
from scipy.stats import uniform

n = 1000
start = 100
width = 500

data_uniform = uniform.rvs(size = n, loc = start, scale=width )
```

```python
ax = sns.distplot(data_uniform,
                 bins = 50,
                kde = True, #density plot option
                 color = 'skyblue',
                 hist_kws = {'linewidth': 1.5, 'alpha':1})

ax.set(xlabel = 'Uniform Distribution', ylabel = 'Frequency')
```

# Normal Distribution (Gaussian Distribution)
1. density curve is symmetrical
2. centered about its mean. Spread is determined by standard deviation. Data near mean are more frequent than others.
3. 68% data in one std, 95%in 2 std.
4. loc  = mean ; scale = std

```python
from scipy.stats import norm

data_norm = norm.rvs(size = 100000, loc = 50, scale = 1)
```

```python
ax = sns.distplot(data_norm,
                 kde  =True,
                 bins = 100,
                 color = 'orange',
                 hist_kws = {'linewidth': 1.5, 'alpha':1} 
                 )

ax.set(xlabel = 'Normal Distribution', ylabel = 'Frequency')
```

# 3. Exponential Dirstribution
1. describes the time between events in a Poisso point process : a process in which events occus continuously and independently at a constant average rate
2. Examples: 
    Length of time between metro arrivals
    length of time between arivals at a gas station
    the life of an air conditioner
3. scale = shape parameter = 1/lambda ; loc = distribution shift; size = no. of variates

```python
from scipy.stats import expon

data_expo = expon.rvs(size = 1000, loc = 0, scale = 1)
```

```python
ax = sns.distplot(data_expo,
                 color = 'green',
                 kde = True,
                 bins = 50,
                 hist_kws = {'linewidth': 1.5, 'alpha':1})
ax.set(xlabel = 'Exponenial distribution', ylabel = 'frequency')
```

# 4. Poisson Distribution
1. number of times an event happened in a time interval. 
2. Example: 
    number of emergency calls recorded at a call centre in a day.
    the number of customers arriving at a salon.
    number of printing errors at each page of the book.
3. avg number of events in an interval = mu= lambda= rate parameter.

NOTE: normal distribution is a limiting case of Poisson distribution with data parameter lambda --> infinity.

```python
from scipy.stats import poisson

data_pois = poisson.rvs(mu = 3, size = 100000)
```

```python
ax = sns.distplot(data_pois,
                 bins = 100,
                 kde = False,
                 color = 'magenta',
                 hist_kws = {'linewidth': 1.5, 'alpha':None})
```

# Subplots of Uniform Distribution Graph

```python
import seaborn as sns
from scipy.stats import uniform
from IPython.core.display import Image
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(10,10)})

import warnings
warnings.filterwarnings('ignore')

sampleSize = [100,1000,10000,100000]
sampleLen = len(sampleSize)
print(sampleLen)

list1 = []
list2 = []
colorArray = ['green','magenta','blue','red']
i = 1

start = 100
width = 1000

f,axes = plt.subplots(1,4)

for i in range(sampleLen+1):
    data_uniform = 'data_uniform' + str(i)
    list1.append(data_uniform)
    
    ax = 'ax'+ str(i)
    list2.append(ax)

for i in range(sampleLen):
    list1[i] = uniform.rvs(size=sampleSize[i], loc = start, scale = width)
    list2[i] = sns.distplot(list1[i],
                bins=sampleSize[i],
                kde=True,
                color=colorArray[i],
                hist_kws={"linewidth": 15,'alpha':1},
                ax=axes[i,])

    list2[i].set(xlabel='Uniform Distribution ', ylabel='Frequency')
```

# Bionomial Distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

import warnings
warnings.filterwarnings('ignore')

sns.set(rc = {'figure.figsize': (5,5)})

f, axes = plt.subplots(2,2)

n=10
p=[0.1,0.3,0.4,0.5]
size = [1000,10000,100000,1000000]
i = 0
k=0
list = []

colorArray = ['red','green','yellow','magenta']

# for i in range(4):
# data_binom = np.random.binomial(n,p,size)

while i<2:
    j=0
    while j<2:
        data_binom = np.random.binomial(n,p[k],size[k])
        ax = sns.distplot(data_binom,
                         kde=True,
                         bins = 50,
                         color = colorArray[k],
                         hist_kws = {'linewidth': 1.5, 'alpha':1},
                         ax = axes[i,j])
        j=j+1
        k = k+1
    i = i+1
```

# Geometric Distribution

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set(rc={'figure.figsize':(10,10)})

f, axes = plt.subplots(2,2)

prob = [0.1,0.5,0.9,0.4]

rv = np.random.RandomState()
k = 0
colorArray = ['blue','green','magenta','yellow']

for i in range(2):
    for j in range(2):
        data_geom = rv.geometric(prob[k],size = 100000)
        ax = sns.distplot(data_geom,
                          kde = True,
                          bins = 50,
                          color = colorArray[k],
                          hist_kws = {'linewidth':1.5, 'alpha':1},
                          ax = axes[i,j]                          
                         )
        ax.set(xlabel = 'Geometric Distribution')
        k = k+1
        j=j+1
    i = i+1
    
```
